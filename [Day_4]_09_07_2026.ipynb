# Feature Description

State - Name of the Indian State.

Year - Year in which the data was recorded.

Installed - Bio-Mass Power - Biomass energy generation capacity.

Installed - Small Hydro Power - Small hydro power generation capacity.

Installed - Solar Power - Installed solar power capacity.

Installed - Wind Power - Installed wind power capacity.

Zone - Geographical zone of India.

Installed - Waste to Energy - Energy generated from waste.

Population - Population of the state.

GSDP (crore INR) - Gross State Domestic Product.

# General Observations

• Population generally increases every year.

• Solar capacity has increased over the years.

• Wind power is higher in coastal states.

• Waste-to-energy capacity is comparatively lower.

• Different renewable sources contribute differently depending on the state.

# Check null values

```python
df.isnull().sum()
```

# Fill null values with mean of cloumn

```python
df.fillna(df.mean(numeric_only=True), inplace=True)
df.isnull().sum()
```

```python
print("Average Solar Power :", df["Installed - Solar Power"].mean())

print("Average Wind Power :", df["Installed - Wind Power"].mean())

print("Average Biomass Power :", df["Installed - Bio-Mass Power"].mean())

print("Average Small Hydro :", df["Installed - Small Hydro Power"].mean())
```
